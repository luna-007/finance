```python
import pandas as pd
import requests
import numpy as np
from scipy import stats
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('sp_500_stocks.csv')
df
```

```python
from secrets import IEX_CLOUD_API_TOKEN

def chunks(lst, n):
    
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
        
symbols = []
symbol_groups = list(chunks(df['Ticker'], 100))

for i in range (len(symbol_groups)):
    symbols.append(','.join(symbol_groups[i]))

for name in symbols:
    # print(name)
    batch_api_url = f'https://sandbox.iexapis.com/stable/stock/market/batch/?types=stats,quote&symbols={name}&token={IEX_CLOUD_API_TOKEN}'
    data = requests.get(batch_api_url).json()
```
